```
import pandas as pd
import numpy as nm
import seaborn as sns
import sklearn as sk
import matplotlib as plt
import matplotlib.pyplot as plt
```

```
df=pd.read_csv("https://raw.githubusercontent.com/SourabhGupta7/House-Sales-in-King-County/master/kc_house_data.csv")
df.head(10)
```

```
df.info()
```

```
df.isnull().sum()
```

```
df.drop(["id","date",],axis=1)
```

```
from sklearn.model_selection import train_test_split
```

```
y=df["price"]
X=df.drop(["price","date"],axis=1)
```

```
X_test,X_train,y_test,y_train=train_test_split(X,y,test_size=0.3,random_state=1)
```

```
from sklearn.linear_model import LinearRegression
```

```
lm=LinearRegression()
```

```
lm.fit(X_train,y_train)
```

```
predictions=lm.predict(X_test)
accuracy=lm.score(X_test,y_test)
print(accuracy)
```
